```python
import pandas as pd
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("sepet.csv", names=['products'], header=None, encoding="ISO-8859-1")
```

```python
data = list(df["products"].apply(lambda x:x.split(';')))
data
```

```python
te = TransactionEncoder()
```

```python
te_data = te.fit(data).transform(data)
df = pd.DataFrame(te_data, columns=te.columns_)
df
```

```python
df1 = apriori(df, min_support=0.02, use_colnames=True, verbose=1)
df1
```

```python
association_rules(df1, metric="confidence", min_threshold=0.6)
```

```python
rules=association_rules(df1, metric="confidence", min_threshold=0.6)
rules[(rules['confidence']>=0.8)&(rules['support']>=0.2)&(rules['lift']>=1.0)]
```
